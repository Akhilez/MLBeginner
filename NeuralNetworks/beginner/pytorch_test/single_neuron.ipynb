```python
import torch
```

Equation to simulate => y = 3x - 2

Steps involved:
 - Get your training data
 - Create the weights and biases
 - Create forward pass graph
 - Create training loop

```python
x_train = torch.tensor([[0.7], [0.5], [1.1]])
y_train = torch.tensor([[0.1], [-0.5], [1.3]])

x_test = torch.tensor([[0.8], [0.9]])
y_test = torch.tensor([[0.4], [0.7]])
```

```python
input_shape = (1,)
num_neurons = 1

weights = torch.rand(input_shape, requires_grad=True)
biases = torch.rand((num_neurons), requires_grad=True)

weights, biases
```

```python
def forward_pass(x, y, w, b):
    star = w.dot(x)
    plus = star + b
    minus = y - plus
    sq = minus ** 2
    return plus, sq
```

```python
learning_rate = 0.5
```

```python
for epoch in range(20):
    print(f"\n----------Epoch = {epoch}")
    losses = []
    for i in range(len(x_train)):
        y_pred, loss = forward_pass(x_train[i], y_train[i], weights, biases)
        
        print(f"\ny_pred = {float(y_pred)}, y_train = {float(y_train[i])}, loss = {float(loss)}")
        losses.append(loss.detach().numpy())

        loss.backward()

        weights = weights - learning_rate * weights.grad
        biases = biases - learning_rate * biases.grad

        weights = weights.detach().requires_grad_()
        biases = biases.detach().requires_grad_()
        
        print(f"New weights = {float(weights), float(biases)}")
        
    total_loss = sum(losses) / len(losses)
    print(f"Total loss = {float(total_loss)}")
```
