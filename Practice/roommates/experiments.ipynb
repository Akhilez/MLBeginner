```python
weighted_data = deepcopy(data)
for key in remapper:
    print(f"Working on {key}")
    column = weighted_data[key]
    for i in range(len(column)):
        value = column[i]
        # print(f"Value = {value}")
        weight = remapper[key][value]
        # print(f"Weight = {weight}")
        weighted_data[key][i] = weight
```

```python
matches = pd.DataFrame(columns=data.index, index=data.index)
matches
```

```python
sliced_weights = weighted_data[remapper.keys()]
sliced_weights
```

```python
for i, user1 in sliced_weights.iterrows():
    for j, user2 in sliced_weights.iterrows():
        if i == j:
            continue
        values = user2[user2 >= user1 - 1]
        matches[j][i] = values
```
